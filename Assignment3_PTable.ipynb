### Main

The main function first open the.json file as mode 'r' by ` with open('','r') as ` and load data from .json file as dict form by `json.load()`. Then close the file and get users' input for diffent operations. Here the exception are wrong type input (automatically detected) and wrong input range (maually raise by `raise`). The opeartion fucntions are in the following cells. 
Option 5 directly return for exit the program.

**Noting:`NewTable='NewPeriodicTable.json'` is just for testing each option function respectively. If your want to modify the original table, Let `NewTable = TableFile`** 

```python
import json

def PeriodicTable(TableFile='PeriodicTableJSON.json',NewTable='NewPeriodicTable.json'): 
    #If your want to modify the original table, Let NewTable = TableFile
    
    with open(TableFile,'r') as f:
        PTable = json.load(f)
    f.close()
    

    Elements   = PTable['elements']
    
    while 1:
        Option = input("Please choose your option from below:\n \
                        Option 1: See all the information that is stored about any element, by entering that element's symbol.\n \
                        Option 2: Choose a property, and see that property for each element in the table. \n \
                        Option 3: Enter a new element. \n  \
                        Option 4: Change the attributes of an element, by entering the element's symbol.\n  \
                        Option 5: Exit the program\n")
                       
        try: 
            Option = int(Option)
            if Option>5 or Option<1:
                       raise ValueError
        except ValueError:
                print("****The input should be an integer between 1 and 5****" )
        else:
           
            if  Option == 1:
                Option_1(Elements)
            elif Option == 2:
                Option_2(Elements)
            elif Option == 3:
                Option_3(Elements,NewTable)
            elif Option == 4:
                Option_4(Elements,NewTable)
            elif Option == 5:
                return
             
              
             
                       
```

### Option 1: See all the information that is stored about any element, by entering that element's symbol.
- Obtain `Symbol` from the keyboard.
- Check its avaiblilty (Sign:`flag`=1 available). Not a existed symbol, activate `ValueError` by raise. Wrong type, automatically activate ValueError. 
- After checking, return all related information of the symbol

```python
def Option_1(Elements):
    
    flag   = 0
    
    while flag == 0:
        try:
            Symbol = input("Please enter your element's symbol:")
            Symbol = str(Symbol)
            for Element in Elements:
                if  Symbol == Element["symbol"]:
                    flag = 1
                    break
            if flag ==0:
                raise ValueError
        except ValueError:
                print("****Please check your symbol input****\n" )
        else:
            print("symbol : %s" %Element["symbol"])
            print("name : %s" %Element["name"])
            print("number : %s" %Element["number"])
            print("row : %s" %Element["xpos"])
            print("column : %s" %Element["ypos"])
```
